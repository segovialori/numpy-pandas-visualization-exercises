```python
#1.  Use matplotlib to plot the following equation:
# y = x^2 - x + 2

#You'll need to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the origin

```

```python
import matplotlib.pyplot as plt
import math

```

```python
x = list(range(0,21))
y = [(num ** 2)-(num + 2) for num in x]
plt.plot(x,y)
plt.annotate('Origin', xy=(0,0), xytext=(2, 100),  arrowprops={'facecolor': 'blue'})
plt.show()
#xytext = (arrow direction, arrow size)

```

```python
#2.  Create and label 4 separate charts for the following equations 
#(choose a range for x that makes sense):
#y = squareroot x
#y = x^3
#y = 2^x
#y = 1/(x+1)

```

```python
#y = squareroot x
x = list(range(0,11))
y1 = [round(math.sqrt(num),2) for num in x]
plt.plot(x,y1)
plt.title('sqrt x')
plt.show()


#y = x^3
y2 = [round(num ** 3,2) for num in x]
plt.plot(x,y2)
plt.title('x ^ 3')
plt.show()

#y = 2^x
y3 = [round(2 ** num,2) for num in x]
plt.plot(x,y3)
plt.title('2 ^ x')

plt.show()
#y = 1/(x+1)
y4 = [round(1/(num + 1),2) for num in x]
plt.plot(x,y4)
plt.title('1/(x + 1)')
plt.show()
```

```python
#3. Combine the figures you created in the last step into one large figure with 4 subplots.
#Figure size
plt.figure(figsize = (12,8))
# plot the first subplot
plt.subplot(221) #plt.subplot(n_rows, n_cols, 1) (nrows, ncolumns, index)
plt.plot(x, y1)
plt.title('sqrt x')

# the second subplot
plt.subplot(222) #plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2)
plt.title('x^3')

# the third subplot
plt.subplot(223)
plt.plot(x, y3)
plt.title('2^x')
# the fourth subplot
plt.subplot(224)
plt.plot(x, y4)
plt.title('1/(x+1)')
plt.show()


```

```python
#4. Combine the figures you created in the last step into 
#one figure where each of the 4 equations has a different color for the points. 
#Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize=(20, 12)) # (width, height)

plt.scatter(x, y1, c='blue')
plt.scatter(x, y2, c='orange')
plt.scatter(x, y3, c='red')
plt.scatter(x, y4, c='green')
plt.plot(x, y1, c='blue', label='$sqrt x$')
plt.plot(x, y2, c='orange', label='$x^3$')
plt.plot(x, y3, c='red', label='$2 ^ x$')
plt.plot(x, y4, c='green', label='$1/(x+1)$')         
         
plt.legend(loc='upper right')
plt.title('X manipulations')
plt.xlabel('$x$')

plt.show()

```
