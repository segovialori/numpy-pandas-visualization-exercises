```python
#Make a new Jupyter notebook named big_o_notation.ipynb

#Title your chart "Big O Notation"
#Label your x axis "Elements"
#Label your y axis "Operations"
#Label your curves or make a legend for the curves
#Use LaTex notation where possible
import matplotlib.pyplot as plt
import math
import pandas as pd

math.log(2)


```

```python
#Curves to graph
x = range(1,4)

# y = 0x + 1 and label the curve "O(1)"
y1 = [0 * num + 1 for num in x]

# y = log(x) and label the curve "O(log n)"
y2 = [math.log(num) for num in x]

# y = x and label the curve "O(n)"
y3 = [num for num in x]

# y = x * log(x) and label it "O(n log n)"
y4 = [num * math.log(num) for num in x]

# y = x ^ 2 and label it "O(n^2)"
y5 = [num ** 2 for num in x]

# y = 2 ^ x and label it "O(2^n)"
y6 = [2 ** num for num in x]

# y = n! and label it "O(n!)"
y7 = [math.factorial(num) for num in x]

# y = n^n and label it (O(n^n))
y8 = [num ** num for num in x]


plt.figure(figsize=(18, 12)) # (width, height)

plt.plot(x, y1, c='forestgreen', label='O(1)')
plt.plot(x, y2, c='limegreen', label='O$\log_{n}$')
plt.plot(x, y3, c='darkgreen', label='O(n)')
plt.plot(x, y4, c='lime', label='O$(n log_{n}$')         
plt.plot(x, y5, c='seagreen', label='O$(n^2)$') 
plt.plot(x, y6, c='mediumseagreen', label='O$(2^n)$')   
plt.plot(x, y7, c='springgreen', label='O(n!)')  
plt.plot(x, y8, c='mediumspringgreen', label='$O(n^n)$')   

plt.legend(loc='upper right')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')


#Bonus Write the code necessary to write your name on a chart. Use box letters.
plt.text(1.0, 25, 'Lori Segovia', fontsize=16, color='grey')

#Bonus: use curves for letters in your name that have curves in them.


plt.show()

```
